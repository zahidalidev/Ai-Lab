```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

import os
df = pd.read_csv('iris.csv')
df.head()

df.isnull().any().any()
df.isna().any().any()

x = df.sepal_length
y = df.petal_width
plt.xlabel('sepal_length')
plt.ylabel('petal_width')
plt.scatter(x,y,color='teal',marker='.')
plt.show()

# import seaborn as sns
# df.drop('Id', axis=1, inplace=True)
# sns.pairplot(df,hue='species')


X = df[['sepal_length','sepal_width','petal_length','petal_width']]
Y = df[['species']]

# unique labels
y_labels = df['species'].unique()

#label encoding
from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(Y)

y_for_plot = pd.DataFrame(y, columns=['Target'])
y_for_plot

# Plotting figuers

plt.figure(figsize=(12,3))
colors = np.array(['yellow', 'green', 'pink'])
iris_targets_legend = np.array(y_labels)

red_patch = mpatches.Patch(color='yellow', label='Setosa')
green_patch = mpatches.Patch(color='green', label='Versicolor')
blue_patch = mpatches.Patch(color='pink', label='Virginica')

plt.subplot(1, 2, 1)
plt.scatter(X['sepal_length'],X['sepal_width'],c= colors[y_for_plot['Target']])
plt.title('Sepal Length vs Sepal Width')
plt.legend(handles=[red_patch, green_patch, blue_patch])
            
        
plt.subplot(1,2,2)
plt.scatter(X['petal_length'], X['petal_width'], c=colors[y_for_plot['Target']])
plt.title('Petal Length vs Petal Width')
plt.legend(handles=[red_patch, green_patch, blue_patch])


from sklearn.cluster import KMeans
import sklearn.metrics as sm

model = KMeans(n_clusters=3)  # 3 clusters known
model.fit(X)

y_pred_ =model.labels_

print("Model Predcition ")
print(y_pred_)

print (model.cluster_centers_)

# plotting clusters
plt.figure(figsize=(12,3))

colors = np.array(['yellow', 'green', 'pink'])

predictedY = np.choose(model.labels_, [1, 0, 2]).astype(np.int64)

plt.subplot(1, 2, 1)
plt.scatter(X['petal_length'], X['petal_width'], c=colors[y_for_plot['Target']])
plt.title('Before classification')
plt.legend(handles=[red_patch, green_patch, blue_patch])

plt.subplot(1, 2, 2)
plt.scatter(X['petal_length'], X['petal_width'], c=colors[predictedY])
plt.title("Model's classification")
plt.legend(handles=[red_patch, green_patch, blue_patch])
```
